```python
#Dataset Preprocessing(removing stop words and html tags and punctuations)

corpus = []
for i in range(0, len(df)):
    review = re.sub('[^a-zA-Z]', ' ', df["para_text"][i])#replace everything excepts aphabets
    review = review.lower()
    review = review.split()
    
    review = [word for word in review if not word in stopwords.words('english')]
    review=" ".join(review)
    corpus.append(review)
print("done")
```

```python
def clean(text):
    review = re.sub('[^a-zA-Z]', ' ',text )#replace everything excepts aphabets
    review = review.lower()
    review = review.split()
    
    review = [word for word in review if not word in stopwords.words('english')]
    review=" ".join(review)
    return review
```

```python
def embedd(corpus):
    feats=[]
    for x in corpus:
        features=nlp(x)
        features=features.vector
        if np.count_nonzero(features)==0:
            return [np.ones(300)]
        feats.append(features)
    return feats
```
